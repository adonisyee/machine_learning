```python
import numpy as np
import matplotlib.pyplot as plt
from imageio import imread

%matplotlib inline 
```

# Part a)

```python
def low_rank_approximation(X, rank):
    # YOUR CODE GOES HERE
    U, Sigma, V = np.linalg.svd(X, full_matrices=False)
    M_hat = np.dot(U[:,:rank], np.dot(np.diag(Sigma[:rank]), V[:rank, :]))   
    return M_hat
```

```python
face = imread("./data/face.jpg")
```

```python
plt.imshow(face, cmap='gray')
plt.title('Original Face')
plt.show()
rank_5_face = low_rank_approximation(face, 5)
plt.imshow(rank_5_face, cmap='gray')
plt.title('Rank 5 Approximation: Face')
plt.show()
rank_20_face = low_rank_approximation(face, 20)
plt.imshow(rank_20_face, cmap='gray')
plt.title('Rank 20 Approximation: Face')
plt.show()
rank_100_face = low_rank_approximation(face, 100)
plt.imshow(rank_100_face, cmap='gray')
plt.title('Rank 100 Approximation: Face')
plt.show()
```

# Part c)

```python
def mse(img1, img2):
    # YOUR CODE GOES HERE 
    return np.mean((img1 - img2)**2)
```

```python
mse_face = []
mse_sky = []
for k in np.arange(1, 101):
    face_k = low_rank_approximation(face, k)
    sky_k = low_rank_approximation(sky, k)
    mse_face.append(mse(face, face_k))
    mse_sky.append(mse(sky, sky_k))
```

```python
plt.plot(np.arange(1, 101), mse_face, label='MSE: Face')
plt.plot(np.arange(1, 101), mse_sky, label='MSE: Sky')
plt.legend()
plt.title('MSE for Face and Sky Image Approximations')
plt.xlabel('Rank')
plt.ylabel('MSE');
```

It becomes difficult for me to distinguish between the low rank approximation for the sky image and the original sky image at a much lower rank than for the low rank approximation for the face image and the original face image. For the face image, it becomes difficult for me to distinguish between the two at rank of about 70, while for sky it becomes difficult for me to distinguish at a rank of about 25. Possible reasons for this are because the face image is more detailed so it requires more information to be able to accurately display it, and also maybe because I am used to looking at and noticing differences in faces more than I am used to finding differences in the sky. Also, we see from the graph in part C that the error is reduced for sky much faster than it is for face, so it makes sense that the sky images look more similar than face images at lower rank.
